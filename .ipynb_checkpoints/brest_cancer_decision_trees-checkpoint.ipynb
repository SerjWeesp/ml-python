```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_curve, roc_auc_score, classification_report
%matplotlib inline
```

```python
SEED = 23
```

```python
def plot_densities(data, names):
    # Set fig size
    plt.rcParams['figure.figsize'] = 20, 20

    # Init figure
    fig, axs = plt.subplots(len(names) - 1, 1)
    fig.suptitle('Features densities for different outcomes 0/1')
    plt.subplots_adjust(left=0.25, right=0.9, bottom=0.1, top=0.95, wspace=0.2, hspace=0.9)

    # Plot densities for outcomes
    for idx, column_name in enumerate(names[:-1]):
        ax = axs[idx]
        data[data['target'] == 0][column_name].plot(kind='density', ax=ax, 
                                                         color="green", legend=True,
                                                         label=f'{column_name} for DEATH_EVENT = 0')
        data[data['target'] == 1][column_name].plot(kind='density', ax=ax, 
                                                         color="red", legend=True,
                                                         label=f'{column_name} for DEATH_EVENT = 1')
        ax.set_xlabel(f'{column_name} values')
        ax.set_title(f'{column_name} density')
        ax.grid('on')

    plt.show()
    plt.rcParams['figure.figsize'] = 20, 10
```

```python
def Winsorization_Method(df, columns,  lower, upper):
    outliers=[]

    for col in columns:
        q1= np.percentile(df[col], lower)
        q2= np.percentile(df[col], upper)
        
        for pos in range(len(df)):
            if df[col].iloc[pos]>q2 or df[col].iloc[pos]<q1:
                outliers.append(pos) 
                
    outliers= set(outliers)                   # remove the duplicates from the outliers
    outliers= list(outliers)
    
    ratio= round(len(outliers)/len(df)*100, 2)                       # Ratio of outliers
    df_win = df.drop(df.index[outliers])    # remove the outliers from the training dataset
    
    return ratio, df_win
```

```python
#load the data

cancer = load_breast_cancer()
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df['target'] = cancer.target
```

```python
#calculating correlation matrix
corr = df.drop('target', axis=1).corr('spearman')

# Generate a mask for the upper triangle
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

fig, ax = plt.subplots(figsize=(12, 12))
sns.heatmap(corr, mask=mask,  annot=corr.round(1))
```

```python
#Show density plot for independend variables divided by target variable's levels
plot_densities(df, df.columns)
```

```python
# Identify pairs of highly correlated variables

high_corr_vars = []
for i in range(len(corr.columns)):
    for j in range(i):
        if corr.iloc[i, j] > 0.8:
            colname_i = corr.columns[i]
            colname_j = corr.columns[j]
            
            # Calculate mean correlation for each variable
            mean_corr_i = corr[colname_i].mean()
            mean_corr_j = corr[colname_j].mean()
            
            # Decide which variable to remove based on mean correlation
            if mean_corr_i > mean_corr_j:
                high_corr_vars.append((colname_i, mean_corr_i))
            else:
                high_corr_vars.append((colname_j, mean_corr_j))

# Remove the variable with the highest mean correlation from each pair
for var, mean_corr in high_corr_vars:
    if var in df.columns:
        print(f"Removing {var} with mean correlation = {mean_corr}")
        del df[var]

print("Remaining columns:", df.columns)
```

```python
num_cols = df.drop('target', axis = 1).columns
```

```python
#cutting outliers

ratio, df_win = Winsorization_Method(df, num_cols, lower=0.03, upper=99.7)
print(f"Ratio of Outliers Detected in the dataset:{ratio}")
```

```python
X_train, X_test,  y_train, y_test = train_test_split(df.drop('target', axis=1), 
                                                    df['target'], test_size=0.1, random_state=SEED, stratify=df['target'])
```

```python
X_train.shape,X_test.shape, y_train.shape, y_test.shape
```

```python
rf = RandomForestClassifier(random_state=SEED)
```

```python
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
y_pred_proba = rf.predict_proba(X_test)[:,1]
```

```python
accuracy_score(y_pred, y_test)
```

```python
roc_auc_score(y_test, y_pred_proba)
```

```python
#feature importance

importances_rf = pd.Series(rf.feature_importances_, index = X_train.columns).sort_values()
importances_rf.plot(kind = 'barh', color = 'lightgreen')
```
